## Population by race 2019-2015 NC per county

```python
url = f"https://api.census.gov/data/2019/pep/charagegroups?get=NAME,POP,RACE&AGEGROUP=5,6,7,8,9,10,11,12,13&DATE_CODE=8,9,10,11,12&DATE_DESC&for=county:*&in=state:37&key={census_api_key}"
response = requests.request("GET", url)
```

```python
df_race = json_to_df(response)
```
